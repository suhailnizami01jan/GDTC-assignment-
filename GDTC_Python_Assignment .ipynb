# Task 1
### Merge the 3 dataset and create 1 view of data.
### You can merge insurance_data.csv and employee_data.csv on
### AGENT_ID You can merge insurance_data.csv and vendor_data.csv on
### VENDOR_ID Note: Use left Outer join as not all claims require Vendor


```python
import pandas as pd
pd.pandas.set_option('display.max_columns',None)
import numpy as np
```

```python
employee_data = pd.read_csv("employee_data.csv")
employee_data.head()
```

```python
print("employee_data dimensions: {}".format(employee_data.shape))
```

```python
insurance_data = pd.read_csv("insurance_data.csv")
insurance_data.head()
```

```python
print("insurance_data dimensions: {}".format(insurance_data.shape))
```

```python
vendor_data = pd.read_csv("vendor_data.csv")
vendor_data.head()
```

```python
print("vendor_data dimensions: {}".format(vendor_data.shape))
```

```python
# Merging of insurance_data.csv and employee_data.csv on AGENT_ID 
# and insurance_data.csv and vendor_data.csv on VENDOR_ID,
# we can merge it individually in pair of 2 but here i raise sub query.
merge_data = pd.merge(pd.merge(insurance_data,employee_data,on="AGENT_ID",how ="left"),vendor_data,on="VENDOR_ID",how="left")
```

```python
merge_data.head()
```

```python
print("merge_data dimensions: {}".format(merge_data.shape))
```
